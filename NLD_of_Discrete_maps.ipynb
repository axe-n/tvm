# The Bifurcation diagram for the  logistic map

```python
import numpy as np
import matplotlib.pyplot as plt
def logistic_map(Xn,r,t):
    z=np.zeros(t)
    for i in range (t):
        y= Xn*(r*(1-Xn))
        Xn=y
        z[i]=Xn
    return z
t=1000
r = np.linspace(3.82,3.85,1000)
for i in r:
    X= logistic_map(np.random.random(),i,t)
    x=X[900:]
    v=np.zeros_like(x)
    v[:]=i
    plt.scatter(v,x,color="black",s=0.2,alpha=0.5)
plt.xlabel("r")
plt.ylabel("x")
plt.title("BIfurcation diagram for logistic map")
plt.show()
```

# liapunov_exponent vs r

```python
import numpy as np
import matplotlib.pyplot as plt

def lyapunov_exponent(r_values, t=1000):
    lyap = np.zeros_like(r_values)
    for idx, r in enumerate(r_values):
        x = 0.5
        le = 0.0
        for i in range(t):
            x = r * x * (1 - x)
            derivative = abs(r * (1 - 2 * x))
            if derivative == 0:
                derivative = 1e-10
            le += np.log(abs(derivative))
        lyap[idx] = le / t
    return lyap

r = np.linspace(3, 4, 1000)
lyap = lyapunov_exponent(r)

plt.plot(r, lyap, color='blue')
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('r')
plt.ylabel('Lyapunov Exponent')
plt.title('Lyapunov Exponent for Logistic Map')
plt.grid(True)
plt.show()
```

# sensitive_dependence on initial condition

```python
def logistic_map(x0, r, n):
    x = np.zeros(n)
    x[0] = x0
    for i in range(1, n):
        x[i] = r * x[i-1] * (1 - x[i-1])
    return x

n = 30
x0_1 = 0.2
x0_2 = 0.201

for a_val in [3.99, 3.29]:
    x1 = logistic_map(x0_1, a_val, n)
    x2 = logistic_map(x0_2, a_val, n)

    plt.figure(figsize=(8, 4))
    plt.plot(x1, label='x = 0.2',color='blue' ,linestyle='--')
    plt.plot(x2, label='x = 0.201',color='green', linestyle='--')
    plt.xlabel('n (iteration)')
    plt.ylabel('x')
    plt.title(f'Logistic Map Iterates for a = {a_val}')
    plt.legend()
    plt.grid(True)
    plt.show()
```
